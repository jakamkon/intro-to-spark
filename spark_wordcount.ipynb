```python
from pyspark.sql import SparkSession
from operator import add

spark = SparkSession.builder.appName("PythonWordCount").getOrCreate()

l = spark.read.text("./worldcount.txt")
l.head(1)
```

```python
l.count()
```

```python
lines = l.rdd
lines
```

```python
lines = lines.map(lambda r: r[0])
lines.take(10)
```

```python
s1 = lines.flatMap(lambda x: x.split(' '))
s1.take(10)
```

```python
s2 = s1.map(lambda x: (x, 1))
s2.take(10)
```

```python
s3 = s2.reduceByKey(add)
s3.take(10)
```

```python
output = s3.collect()

for (word, count) in output:
    print("%s: %i" % (word, count))
    
spark.stop()
```
