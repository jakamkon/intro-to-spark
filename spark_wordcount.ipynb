```python
from pyspark.sql import SparkSession
from operator import add

spark = SparkSession.builder.appName("PythonWordCount").getOrCreate()
```

```python
s1 = spark.read.text("./worldcount.txt").rdd.map(lambda r: r[0])
s1.take(10)
```

```python
s2 = s1.flatMap(lambda x: x.split(' '))
s2.take(10)
```

```python
s3 = s2.map(lambda x: (x, 1))
s3.take(10)
```

```python
s4 = s3.reduceByKey(add)
s4.take(10)
```

```python
output = s4.collect()

for (word, count) in output:
    print("%s: %i" % (word, count))
    
spark.stop()
```
